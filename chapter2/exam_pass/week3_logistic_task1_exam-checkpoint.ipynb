# 逻辑回归预测考试通过

**任务:**

基于examdata.csv数据，建立逻辑回归模型
预测Exam1 = 75, Exam2 = 60时，该同学在Exam3是 passed or failed;
建立二阶边界，提高模型准确度

```python
# Load data
import pandas as pd
import numpy as np
data = pd.read_csv('examdata.csv')
data.head()
```

```python
# visualize data
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize=(6, 4))
plt.scatter(data.loc[:, 'Exam1'], data.loc[:, 'Exam2'])
plt.title('Exam1 - Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.show()
```

```python
# add label mask
mask = data.loc[:, 'Pass'] == 1
print(~mask)
print(mask)
```

```python
fig2 = plt.figure()
passed = plt.scatter(data.loc[:, 'Exam1'][mask], data.loc[:, 'Exam2'][mask])
failed = plt.scatter(data.loc[:, 'Exam1'][~mask], data.loc[:, 'Exam2'][~mask])
plt.title('Exam1 - Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed,failed),('passed, failed'))
plt.show()
```

```python
# Define X, y
X = data.drop(['Pass'], axis=1)
y = data.loc[:, 'Pass']
X1 = data.loc[:, 'Exam1']
X2 = data.loc[:, 'Exam2']
X1.head()
```

```python
print(X.shape,y.shape)
```

```python
# train the model
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X, y)
```

```python
#show the predicted result and its accuracy
y_predict = LR.predict(X)
print(y_predict)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_predict)
print(accuracy)
```

```python
# test a data
y_test = LR.predict([[70, 65]])
print('passed' if y_test == 1 else 'failed')
```

```python
# paramater b  截距
LR.intercept_
```

### 边界函数  decision boundary

边界函数： $\theta_0 + \theta_1 X_1 + \theta_2 X_2 = 0$

二阶边界函数：$\theta_0 + \theta_1 X_1 + \theta_2 X_2+ \theta_3 X_1^2 + \theta_4 X_2^2 + \theta_5 X_1 X_2 = 0$



```python
#  从训练好的模型中 获取参数
theta0 = LR.intercept_  # b
theta1 = LR.coef_[0][0]  # w1
theta2 = LR.coef_[0][1]  # w2
print(f'theta0 = {theta0}, theta1 = {theta1}, theta2 = {theta2}')
```

```python
# b + w1*x1 + w2*x2 = 0
# x2 = -(b + w1*x1)/w2
X2_new = -(theta0 + theta1 * X1) / theta2
print(X2_new)
```

```python
fig3 = plt.figure()
passed = plt.scatter(data.loc[:, 'Exam1'][mask], data.loc[:, 'Exam2'][mask])
failed = plt.scatter(data.loc[:, 'Exam1'][~mask], data.loc[:, 'Exam2'][~mask])
plt.plot(X1, X2_new)  # 画出decision boundary
plt.title('Exam1 - Exam2')
plt.xlabel('Exam1')
plt.ylabel('Exam2')
plt.legend((passed,failed),('passed, failed'))
plt.show()
```

### 采用二阶边界函数 来训练新的模型
二阶边界函数：$\theta_0 + \theta_1 X_1 + \theta_2 X_2+ \theta_3 X_1^2 + \theta_4 X_2^2 + \theta_5 X_1 X_2 = 0$

```python

X1_2 = X1 * X1
X2_2 = X2 * X2
X1_X2 = X1 * X2
```

```python
X_new = pd.DataFrame({'X1': X1, 'X2': X2,'X1_2': X1_2,  'X2_2': X2_2, 'X1_X2': X1_X2, })
print(X_new.head())
```

```python
LR2 = LogisticRegression()
LR2.fit(X_new, y)
```

```python
y2_predict = LR2.predict(X_new)
accuracy2 = accuracy_score(y,y2_predict)
print(accuracy2)
```

```python
X1_new = X1.sort_values()
print(X1,X1_new)
```

边界函数： $\theta_0 + \theta_1 X_1 + \theta_2 X_2 = 0$

二阶边界函数：$\theta_0 + \theta_1 X_1 + \theta_2 X_2+ \theta_3 X_1^2 + \theta_4 X_2^2 + \theta_5 X_1 X_2 = 0$

$a x^2 + b x + c=0: x1 = (-b+(b^2-4ac)^.5)/2a,x1 = (-b-(b^2-4ac)^.5)/2a$

$\theta_4 X_2^2 + (\theta_5 X_1++ \theta_2) X_2 + (\theta_0 + \theta_1 X_1 + \theta_3 X_1^2)=0$

```python
# 画出新的二阶决策边界
theta0 = LR2.intercept_
theta1, theta2, theta3, theta4, theta5 = LR2.coef_[0][0], LR2.coef_[0][1], LR2.coef_[0][2], LR2.coef_[0][3], LR2.coef_[0][4]
print(theta0, theta1, theta2, theta3, theta4, theta5)
a = theta4
b = theta5 * X1_new + theta2
c = theta0 + theta1 * X1_new + theta3 * (X1_new * X1_new)

X2_new_boundary = (-b + np.sqrt(b * b - 4 * a * c)) / (2 * a)

print(X2_new_boundary)
```
